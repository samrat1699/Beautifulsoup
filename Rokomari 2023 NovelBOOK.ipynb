#### Final work

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

final = pd.DataFrame()

for page in range(1, 15):
    webpage = requests.get("https://www.rokomari.com/book/category/3402/boimela-2023-novel?page={}".format(page)).text
    soup = BeautifulSoup(webpage, 'html.parser')
    
    company = soup.find_all('div', class_='book-list-wrapper')
    books = []
    booksA = []
    Price = []
    disPrice = []
    rating = []
    
    for i in company:
        book_title = i.find('p', class_='book-title')
        books.append(book_title.text.strip() if book_title else np.nan)
        
        book_author = i.find('p', class_='book-author')
        booksA.append(book_author.text.strip() if book_author else np.nan)
        
        original_price = i.find('strike', class_='original-price pl-2')
        Price.append(original_price.text.strip() if original_price else np.nan)
        
        discount_price = i.find('span', class_='')
        disPrice.append(discount_price.text.strip() if discount_price else np.nan)
        
        book_rating = i.find('span', class_='text-secondary')
        rating.append(book_rating.text.strip() if book_rating else np.nan)
            
    df = pd.DataFrame({'books': books,
                       'booksA': booksA,
                       'Price': Price,
                       'disPrice': disPrice,
                       'rating': rating
                      })
    
    final = final.append(df, ignore_index=True)

final.to_csv("2023bookfair_novel.csv", index=False)
```
